```python
import pandas as pd
```

```python
quadrant_1 = pd.read_csv("quadrant 1.csv")
quadrant_2 = pd.read_csv("quadrant 2.csv")
quadrant_3 = pd.read_csv("quadrant 3.csv")
quadrant_4 = pd.read_csv("quadrant 4.csv")
```

```python
# create a new dataframe for merged tables
all_quadrants = pd.DataFrame()
all_quadrants = quadrant_1[["X X-Ray Coordinate", "Y X-Ray Coordinate",
                            "Z X-Ray Coordinate", "X Pixel Index",
                            "Y Pixel Index"]]
```

```python
index_to_coordinate = {0:-2, 1:-1, 2:0, 3:1, 4:2}
coordinate_to_index = {-2:0, -1:1, 0:2, 1:3, 2:4}
```

# Add Data for Quadrant 1

```python
all_quadrants["Charge Fraction (0° rotation)"] = quadrant_1["Charge Fraction"]
all_quadrants
```

# Add Data for Quadrant 2

```python
quadrant_2["Original X X-Ray"] = quadrant_2["Y X-Ray Coordinate"]
quadrant_2["Original Y X-Ray"] = -quadrant_2["X X-Ray Coordinate"]
quadrant_2["Original Z X-Ray"] = quadrant_2["Z X-Ray Coordinate"]

quadrant_2
```

```python
quadrant_2["X Pixel Coordinate"] = quadrant_2["X Pixel Index"]
quadrant_2 = quadrant_2.replace({"X Pixel Coordinate":index_to_coordinate})
quadrant_2["Y Pixel Coordinate"] = quadrant_2["Y Pixel Index"]
quadrant_2 = quadrant_2.replace({"Y Pixel Coordinate":index_to_coordinate})

quadrant_2["Original X Pixel Coordinate"] = quadrant_2["Y Pixel Coordinate"]
quadrant_2["Original Y Pixel Coordinate"] = -quadrant_2["X Pixel Coordinate"]

quadrant_2["Original X Pixel Index"] = quadrant_2["Original X Pixel Coordinate"]
quadrant_2["Original Y Pixel Index"] = quadrant_2["Original Y Pixel Coordinate"]

quadrant_2 = quadrant_2.replace({"Original X Pixel Index":coordinate_to_index})
quadrant_2 = quadrant_2.replace({"Original Y Pixel Index":coordinate_to_index})

quadrant_2
```

```python
quadrant_2 = quadrant_2.drop(columns=["X X-Ray Coordinate", "Y X-Ray Coordinate", "Z X-Ray Coordinate",
                                      "X Pixel Index", "Y Pixel Index", "X Pixel Coordinate", "Y Pixel Coordinate",
                                      "Original X Pixel Coordinate", "Original Y Pixel Coordinate"])
quadrant_2
```

```python
all_quadrants = pd.merge(all_quadrants, quadrant_2,  how="left", 
                         left_on=["X X-Ray Coordinate", "Y X-Ray Coordinate", "Z X-Ray Coordinate", "X Pixel Index", "Y Pixel Index"], 
                         right_on = ["Original X X-Ray", "Original Y X-Ray", "Original Z X-Ray", "Original X Pixel Index", "Original Y Pixel Index"])
all_quadrants = all_quadrants.drop(columns=["Original X X-Ray", "Original Y X-Ray", "Original Z X-Ray", "Original X Pixel Index", "Original Y Pixel Index"])
all_quadrants = all_quadrants.rename(columns={"Charge Fraction": "Charge Fraction (90° rotation)"})
all_quadrants 
```
